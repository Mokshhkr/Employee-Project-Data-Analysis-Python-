##Task1: Create Dataframes and save as csv files

```python
import pandas as pd
import numpy as np
```

##Creating Employee Dataframe

```python
employee_df = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005'],
    'Name': ['John Alter','Alice Luxumberg','Tom Sabestine','Nina Adgra','Amy Johny'],
    'Gender': ['M','F','M','F','F'],
    'City': ['Paris','London','Berlin','Newyork','Madrid'],
    'Age': [25,27,29,31,30]
})

employee_df 
```

##save employee CSV in separate cell

```python
employee_df.to_csv('Employee.csv', index=False)
```

##Creating Seniority dataframe

```python
seniority_df = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005'],
    'Designation Level': [2,2,3,2,3]
})

seniority_df
```

##Saving Seniority Level DataFrame

```python
seniority_df.to_csv('Seniority.csv', index=False)
```

##Creating Project Dataframe

```python
project_df = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005','A002','A005','A003',
           'A001','A003','A001','A004','A004','A005'],
    'Project': ['Project 1','Project 2','Project 3','Project 4','Project 5',
                'Project 6','Project 7','Project 8','Project 9','Project 10',
                'Project 11','Project 12','Project 13','Project 14'],
    'Cost': [1002000,2000000,4500000,5500000,np.nan,680000,400000,
             350000,np.nan,300000,2000000,1000000,3000000,200000],
    'Status': ['Finished','Ongoing','Finished','Ongoing','Finished','Failed',
               'Finished','Failed','Ongoing','Finished','Failed','Ongoing',
               'Finished','Finished']
})

project_df
```

##Saving Project Dataframe

```python
project_df.to_csv('Project.csv', index=False)
```

##Task2: Fill missing cost using running Average(using 'For' Loop )

```python
# Reading Project CSV
project_df = pd.read_csv('Project.csv')

# Calculating running average
cost_sum = 0
count = 0

for i in range(len(project_df)):
    if not pd.isna(project_df.loc[i, 'Cost']):
        cost_sum += project_df.loc[i, 'Cost']
        count += 1
    else:
        project_df.loc[i, 'Cost'] = cost_sum / count

# Check result
project_df
```

##Task3: Split Employee name into First name and Last name

```python
# Reading Employee CSV
employee_df = pd.read_csv('Employee.csv')

# Splitting Name column
employee_df[['First Name','Last Name']] = employee_df['Name'].str.split(' ', expand=True)

# Dropping old Name column
employee_df.drop(columns=['Name'], inplace=True)

employee_df
```

##Task4: Final Dataframe

```python
# Reading all CSVs
seniority_df = pd.read_csv('Seniority.csv')

# Merge Employee + Seniority
final_df = pd.merge(employee_df, seniority_df, on='ID', how='inner')

# Merge with Project DataFrame
final_df = pd.merge(final_df, project_df, on='ID', how='inner')

final_df
```

##Task 5:Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
cost only to employees who have finished the projects.

```python
# Adding Bonus column
final_df['Bonus'] = np.where(final_df['Status'] == 'Finished',
                             final_df['Cost'] * 0.05,
                             0)

final_df
```

##Task6:Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

```python
# Demoting designation for failed projects
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1

# Removing employees with designation level > 4
final_df = final_df[final_df['Designation Level'] <= 4]

final_df
```

Task7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# Adding prefix to First Name
final_df['First Name'] = np.where(final_df['Gender'] == 'M',
                                  'Mr. ' + final_df['First Name'],
                                  'Mrs. ' + final_df['First Name'])

# Dropping Gender column
final_df.drop(columns=['Gender'], inplace=True)

final_df
```

##Task 8: Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

```python
# Promoting designation for age > 29
final_df.loc[final_df['Age'] > 29, 'Designation Level'] -= 1

final_df
```

##Task9:Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost

```python
# Grouping total cost per employee
total_cost_df = final_df.groupby(['ID','First Name'])['Cost'].sum().reset_index()

# Renaming column
total_cost_df.rename(columns={'Cost':'Total Cost'}, inplace=True)

total_cost_df
```
